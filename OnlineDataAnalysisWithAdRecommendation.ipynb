```python
import warnings
warnings.filterwarnings("ignore")
!pip install google-generativeai

import pandas as pd
import seaborn as sns
import numpy as np
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
from google.colab import drive

# Mount Google Drive
drive.mount('/content/drive')

# Load dataset
df = pd.read_csv('/content/drive/MyDrive/OnlineRetail.csv', encoding='latin1')

# Convert date column
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])

print(df.head())
print(df.info())

# -----------------------------
# DATA CLEANING
# -----------------------------

# remove rows without CustomerID
df = df.dropna(subset=['CustomerID'])

# remove returned items
df = df[df['Quantity'] > 0]

# convert customerID to string
df['CustomerID'] = df['CustomerID'].astype(str)

print("Dataset shape after cleaning:", df.shape)

# -----------------------------
# MARKET BASKET ANALYSIS
# -----------------------------

# Create transaction matrix
basket = df.groupby(['InvoiceNo','Description'])['Quantity'].sum().unstack().fillna(0)

# Convert quantities to binary (purchased or not)
basket = (basket > 0).astype(int)

# Keep only most frequent products
top_products = basket.sum().sort_values(ascending=False).head(500).index
basket = basket[top_products]

print("Basket shape:", basket.shape)
print(basket.head())

# Find frequent itemsets
frequent_items = apriori(basket, min_support=0.02, use_colnames=True)

# Generate association rules
rules = association_rules(frequent_items, metric="confidence", min_threshold=0.3)

print("Frequent itemsets:", len(frequent_items))
print("Rules:", len(rules))

print(rules[['antecedents','consequents','support','confidence','lift']])

# -----------------------------
# CUSTOMER SEGMENTATION
# -----------------------------

# Create TotalPrice column
df['TotalPrice'] = df['Quantity'] * df['UnitPrice']

# Reference date (latest transaction)
reference_date = df['InvoiceDate'].max()

# Aggregate customer features
customer_data = df.groupby('CustomerID').agg({
    'InvoiceNo':'nunique',      # number of orders
    'Quantity':'sum',           # total items bought
    'TotalPrice':'sum',         # total spending
    'InvoiceDate':'max'         # last purchase date
})

# Rename columns
customer_data.columns = ['TotalOrders','TotalItems','TotalSpending','LastPurchaseDate']

# Calculate Recency
customer_data['Recency'] = (reference_date - customer_data['LastPurchaseDate']).dt.days

# Calculate Average Basket Size
customer_data['AverageBasketSize'] = customer_data['TotalItems'] / customer_data['TotalOrders']

# Drop LastPurchaseDate
customer_data = customer_data.drop(columns=['LastPurchaseDate'])

print(customer_data.head())

# -----------------------------
# CLUSTERING
# -----------------------------

customer_data['TotalSpending'] = np.log1p(customer_data['TotalSpending'])

features = customer_data[['Recency','TotalOrders','TotalItems','TotalSpending','AverageBasketSize']]

# scale features
scaler = StandardScaler()
scaled_data = scaler.fit_transform(features)

# KMeans clustering
kmeans = KMeans(n_clusters=3, random_state=42)

customer_data['Cluster'] = kmeans.fit_predict(scaled_data)

print(customer_data.head())

print("\nCustomers per cluster:")
print(customer_data['Cluster'].value_counts())

# -----------------------------
# VISUALIZATION
# -----------------------------

plt.figure(figsize=(8,6))

plt.scatter(customer_data['TotalOrders'],
            customer_data['TotalSpending'],
            c=customer_data['Cluster'],
            cmap='viridis')

plt.xlabel("Total Orders (Frequency)")
plt.ylabel("Total Spending (Monetary)")
plt.title("Customer Segmentation based on RFM")

plt.show()

print("\n")
cluster_labels = {
    0: "Occasional Customers",
    1: "Regular Customers",
    2: "VIP Customers"
}

# Map cluster numbers to labels
customer_data['CustomerType'] = customer_data['Cluster'].map(cluster_labels)

# Plot using labels instead of numbers
customer_data['CustomerType'].value_counts().plot(
    kind='bar',
    figsize=(6,4)
)

plt.title("Number of Customers in Each Segment")
plt.xlabel("Customer Segment")
plt.ylabel("Number of Customers")

plt.show()

print("\n")
sns.pairplot(customer_data,
             vars=['Recency','TotalOrders','TotalSpending'],
             hue='Cluster')

plt.show()
print("\n")

cluster_profile = customer_data.groupby('Cluster')[['Recency','TotalOrders','TotalSpending']].mean()
print(cluster_profile)
cluster_profile.plot(
    kind='bar',
    figsize=(8,6)
)

plt.title("Customer Segment Profile (Average RFM Values)")
plt.xlabel("Customer Cluster")
plt.ylabel("Average Value")

plt.xticks(rotation=0)

plt.show()
# -----------------------------
# DATASET SUMMARY
# -----------------------------

print("\nDataset Summary")
print("Total transactions:", df['InvoiceNo'].nunique())
print("Total products:", df['Description'].nunique())
print("Total customers:", df['CustomerID'].nunique())

print("\nTop Selling Products")
print(df['Description'].value_counts().head(10))
```

```python
def recommend(product):

    product = product.lower()

    matches = rules[rules['antecedents'].apply(
        lambda x: any(
            product in item.lower() or item.lower() in product
            for item in x
        )
    )]

    # 🔥 fallback: search in consequents too
    if matches.empty:
        matches = rules[rules['consequents'].apply(
            lambda x: any(
                product in item.lower() or item.lower() in product
                for item in x
            )
        )]

    if matches.empty:
        return []

    matches = matches.sort_values(by="confidence", ascending=False)

    results = []

    for _, row in matches.head(5).iterrows():
        results.append({
            "bought": list(row['antecedents']),
            "recommend": list(row['consequents']),
            "confidence": round(row['confidence'],2),
            "lift": round(row['lift'],2)
        })

    return results
```

```python
import os
import random
from PIL import Image
from IPython.display import display

BASE_PATH = "/content/drive/MyDrive/ads_dataset"

def get_category(product):

    product = product.lower()

    if any(word in product for word in ["mug", "tea", "cup", "kitchen"]):
        return "kitchen_gifts"

    elif any(word in product for word in ["candle", "holder", "decor", "lantern"]):
        return "home_decor"

    elif any(word in product for word in ["bag", "box", "wrap"]):
        return "gift_wrapping"

    elif any(word in product for word in ["toy", "doll"]):
        return "toys"

    elif any(word in product for word in ["christmas", "xmas"]):
        return "seasonal"

    elif any(word in product for word in ["pen", "notebook"]):
        return "stationery"

    else:
        return "misc"


def get_ad_image(product):

    category = get_category(product)

    folder = os.path.join(BASE_PATH, category)

    if not os.path.exists(folder):
        return None

    images = os.listdir(folder)

    if len(images) == 0:
        return None

    img_path = os.path.join(folder, random.choice(images))

    return img_path
```

```python
def get_product_stats(product):

    data = df[df['Description'] == product]

    total_purchases = data['InvoiceNo'].nunique()
    total_quantity = data['Quantity'].sum()
    total_revenue = (data['Quantity'] * data['UnitPrice']).sum()
    unique_customers = data['CustomerID'].nunique()

    return {
        "transactions": total_purchases,
        "quantity_sold": int(total_quantity),
        "revenue": round(total_revenue, 2),
        "customers": unique_customers
    }

```

```python
import warnings
warnings.filterwarnings("ignore")
import ipywidgets as widgets
from IPython.display import display, clear_output
from IPython.display import Image as IPImage
search_history = []
search_box = widgets.Text(
    placeholder="Type your search...",
    description="Search:",
    layout=widgets.Layout(width="400px")
)
history_dropdown = widgets.Dropdown(
    options=[],
    description="History:",
    layout=widgets.Layout(width="400px")
)

recommend_button = widgets.Button(
    description="Get Recommendations",
    button_style="success"
)


output = widgets.Output()


def on_button_clicked(b):

    with output:
        clear_output()

        query = search_box.value.strip()

        if query == "":
            print("Please type something.")
            return

        # CALL YOUR EXISTING FUNCTION
        results = recommend(query)

        if not results:
            print("No recommendations found. Try simpler keywords like 'tea', 'bag', 'candle'")
            return

        print(f"Recommendations for: {query}\n")

        from IPython.display import display

        for i, r in enumerate(results, 1):
            if isinstance(r, dict):
                recommended_products = r["recommend"]

                for product in recommended_products:

                    print(f"{i}. Recommended product: {product}")
                    print(f"Confidence: {r['confidence']} | Lift: {r['lift']}")

                    # get ad image
                    img_path = get_ad_image(product)

                    if img_path:
                        print(f"Ad Category: {get_category(product)}")
                        display(IPImage(filename=img_path, width=200))
                    else:
                        print("No ad available")

                    # Display product statistics for the recommended product
                    stats = get_product_stats(product)
                    print(f"Transactions: {stats['transactions']}")
                    print(f"Units Sold: {stats['quantity_sold']}")
                    print(f"Revenue: £{stats['revenue']}")
                    print(f"Unique Customers: {stats['customers']}")

                    print("\n")  # spacing between results

                    # store search history
                    if query not in search_history:
                        search_history.append(query)

                    history_dropdown.options = search_history
            else:
                print(f"Skipping unexpected recommendation format: {r}")


recommend_button.on_click(on_button_clicked)
def fill_from_history(change):
    if change["new"]:
        search_box.value = change["new"]

history_dropdown.observe(fill_from_history, names="value")
display(search_box)
display(history_dropdown)
display(recommend_button)
display(output)
```

### Inter-Cluster Common Products

While the previous section showed the top products *within* each cluster (intra-cluster), this section identifies products that are common across the top lists of *multiple* clusters (inter-cluster). This can highlight universally popular items or products that appeal to a broader range of customer segments.

```python
from collections import defaultdict

# Extract top product names for each cluster
cluster_top_products = defaultdict(list)
for (cluster_id, _, product_desc), quantity in top_cluster_products.items():
    cluster_top_products[cluster_id].append(product_desc)

# Find products that appear in more than one cluster's top list
all_top_products = []
for products_list in cluster_top_products.values():
    all_top_products.extend(products_list)

# Count occurrences of each product
product_counts = pd.Series(all_top_products).value_counts()

# Filter for products appearing in more than one cluster's top list
inter_cluster_common_products = product_counts[product_counts > 1]

print("Products common in the top lists of multiple clusters:")
if not inter_cluster_common_products.empty:
    display(inter_cluster_common_products)
else:
    print("No products found to be common across the top lists of multiple clusters.")
```
